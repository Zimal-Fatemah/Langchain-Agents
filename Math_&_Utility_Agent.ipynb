```python
import os
from dotenv import load_dotenv
from google.colab import userdata
#Setting up my API KEY in google colab.
os.environ["API_KEY"] = userdata.get('API_KEY')
my_key = userdata.get('API_KEY')

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import AgentExecutor, create_react_agent
from langchain.tools import Tool
from langchain import hub

#initializing my brain
llm = ChatGoogleGenerativeAI(
model="gemini-2.5-flash",
temperature=0
)

#making the calculator function
def calculator(expression):
  try:
    return str(eval(expression))
  except:
    return "Invalid expression"

#making the text reverso function
def reverse_text(text):
   return text[::-1]

#initializing the tools
tools = [
 Tool(
   name="Calculator",
   func=calculator,
   description="used to solve kind of math problems"
),
 Tool(
     name="Text Reversor",
     func=reverse_text,
     description="used to reverse text"
 )
]

#using the prebuilt REACT prompt, it instructs my agent to reason first(think) and then act accordingly(by using given tools if neccessary)
prompt = hub.pull("hwchase17/react")

#initializing my agent
agent = create_react_agent(llm,tools,prompt)
agent_executor = AgentExecutor(
agent=agent,
tools=tools,
verbose=False
)

#voila, its ready, query the questions or you can ask for any text to reverse, xoxo:)
query=input("Whats on your mind?")
result = agent_executor.invoke({"input":query})
print(result["output"])
```
